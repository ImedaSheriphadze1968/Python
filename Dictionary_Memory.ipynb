```python
import pickle

def сохранить_словарь(словарь, имя_файла):
    with open(имя_файла, 'wb') as файл:
        pickle.dump(словарь, файл)
    print("Словарь успешно сохранён!")

def загрузить_словарь(имя_файла):
    try:
        with open(имя_файла, 'rb') as файл:
            словарь = pickle.load(файл)
        print("Словарь успешно загружен!")
        return словарь
    except FileNotFoundError:
        print("Файл не найден. Создаём новый словарь.")
        return {}

def показать_словарь(словарь):
    if словарь:
        print("Содержимое словаря:")
        for ключ, значение in словарь.items():
            print(f"{ключ} — {значение}")
    else:
        print("Словарь пуст.")

def main():
    имя_файла = 'мой_словарь.pkl'
    словарь = загрузить_словарь(имя_файла)

    while True:
        print("\nМеню:")
        print("1 — Добавить новую пару слово-перевод")
        print("2 — Показать словарь")
        print("3 — Сохранить и выйти")

        выбор = input("Введите номер действия: ")

        if выбор == '1':
            слово = input("Введите слово: ")
            перевод = input("Введите перевод: ")
            словарь[слово] = перевод
            print(f"Пара '{слово} — {перевод}' добавлена.")
        elif выбор == '2':
            показать_словарь(словарь)
        elif выбор == '3':
            сохранить_словарь(словарь, имя_файла)
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
