```python
# üêº –ó–∞–≥–∞–¥–æ—á–Ω—ã–π –ü–∞–Ω–¥–∞–±–æ—Ç

import random

–æ—Ç–≤–µ—Ç—ã_–ø–∞–Ω–¥—ã = [
    "–ü–æ–∫–æ–π –ø—Ä–∏—Ö–æ–¥–∏—Ç –∫ —Ç–æ–º—É, –∫—Ç–æ –µ—Å—Ç –±–∞–º–±—É–∫ üçÉ",
    "–ò–Ω–æ–≥–¥–∞ –º–æ–ª—á–∞–Ω–∏–µ ‚Äî –ª—É—á—à–∏–π –æ—Ç–≤–µ—Ç üßò‚Äç‚ôÇÔ∏è",
    "–î–∞–∂–µ –ø–∞–Ω–¥–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–∏–ª—å–Ω–æ–π, –∫–æ–≥–¥–∞ –Ω–∞–¥–æ üêæ",
    "–°–ø—Ä–æ—Å–∏ —Å–Ω–æ–≤–∞, –Ω–æ —Å–Ω–∞—á–∞–ª–∞ —Å—ä–µ—à—å —á—Ç–æ-–Ω–∏–±—É–¥—å –≤–∫—É—Å–Ω–æ–µ ü•¢",
    "–ö–∞–∂–¥–æ–µ —É—Ç—Ä–æ ‚Äî —ç—Ç–æ —à–∞–Ω—Å —Å—Ç–∞—Ç—å –ª—É—á—à–µ üåÖ"
]

print("üêº –ü—Ä–∏–≤–µ—Ç! –Ø ‚Äî –ü–∞–Ω–¥–∞–±–æ—Ç. –ó–∞–¥–∞–π –º–Ω–µ –≤–æ–ø—Ä–æ—Å –∏–ª–∏ –Ω–∞–ø–∏—à–∏ '–≤—ã—Ö–æ–¥' –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è.\n")

while True:
    –≤–æ–ø—Ä–æ—Å = input("‚ùì –¢—ã: ")
    if –≤–æ–ø—Ä–æ—Å.lower() in ["–≤—ã—Ö–æ–¥", "quit", "exit"]:
        print("üêº –ü–∞–Ω–¥–∞ –≥–æ–≤–æ—Ä–∏—Ç: –î–æ –Ω–æ–≤—ã—Ö –≤—Å—Ç—Ä–µ—á! üçÉ")
        break
    else:
        –æ—Ç–≤–µ—Ç = random.choice(–æ—Ç–≤–µ—Ç—ã_–ø–∞–Ω–¥—ã)
        print("üß† –ü–∞–Ω–¥–∞: " + –æ—Ç–≤–µ—Ç + "\n")
```
