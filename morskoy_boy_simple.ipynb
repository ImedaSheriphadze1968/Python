```python
# morskoy_boy_simple.py
# Простая игра "Морской бой" для одного игрока (минималистичный вариант)
# Русский язык, понятный и аккуратный вывод

def print_board(board):
    print("  " + " ".join(str(i) for i in range(len(board[0]))))
    for idx, row in enumerate(board):
        print(str(idx) + " " + " ".join(row))

def create_board(size):
    return [["~"] * size for _ in range(size)]

def place_ship(board, ship_coords):
    for x, y in ship_coords:
        board[x][y] = "■"

def check_hit(ship_coords, x, y):
    return (x, y) in ship_coords

def mark_hit(board, x, y, hit):
    board[x][y] = "X" if hit else "O"

def main():
    size = 5
    board = create_board(size)
    ship_coords = [(1,1), (1,2), (1,3)]  # простой корабль из 3 палуб

    print("Добро пожаловать в игру 'Морской бой'!")
    print("Попробуйте потопить корабль из 3 палуб на поле 5x5.")
    print("Введите координаты выстрела (строка и столбец через пробел).")
    print()

    shots = 0
    hits = 0
    total_ship_cells = len(ship_coords)

    while hits < total_ship_cells:
        print_board(board)
        try:
            x, y = map(int, input("Ваш выстрел (строка столбец): ").split())
            if not (0 <= x < size and 0 <= y < size):
                print("Координаты вне поля. Попробуйте снова.")
                continue
        except:
            print("Ошибка ввода. Введите два числа через пробел.")
            continue

        shots += 1
        if check_hit(ship_coords, x, y):
            if board[x][y] == "X":
                print("Вы уже сюда стреляли — попадание!")
            else:
                print("Попадание!")
                mark_hit(board, x, y, True)
                hits += 1
        else:
            if board[x][y] == "O":
                print("Вы уже сюда стреляли — промах!")
            else:
                print("Промах!")
                mark_hit(board, x, y, False)

        print()

    print_board(board)
    print(f"Поздравляем! Корабль потоплен за {shots} выстрелов.")

if __name__ == "__main__":
    main()
```
